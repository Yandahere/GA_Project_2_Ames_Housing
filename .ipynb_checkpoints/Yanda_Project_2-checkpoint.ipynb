## Background
Real Sky is a real estate development company situated in the US. However, the company has been steadily losing profit on the houses they have been building. They are competing with other real estate companies such as Blue Note that uses traditional methods of engineering new development plans. Real Sky has approached us, their in-house Data Scientist team, to create a machine learning model to give them the competitive edge against Blue Note.

### Dataset
Data set contains information from the Ames Assessor’s Office used in computing assessed values for individual residential properties sold in Ames, IA from 2006 to 2010. The dataset is then seperated into a `train.csv` and a `test.csv`.

# Problem Statement:
## What *core features* should Real Sky, focus on to *increase the sale price of homes* for a higher profit for their next development project?

To solve this problem statement, we will be employing a regression model to pinpoint predictive core variables that influence saleprices. We will also be ultilizing RMSE.

We will fill for missing values in catergorical features via mode values and fill the missing values for numerical features with median values.

Let's create a function to fill the NA values with mode or median.

```python
def replace_missing_mode(df, col):
    mode = df.mode()[col][0]
    df.fillna({col: mode}, inplace=True)
```

```python
def replace_missing_median (df, col):
    med = df[col].median()
    df[col].fillna(med, inplace=True)
```

```python
for col in train_df[n_below_fifty]:
    replace_missing_median(train_df, col)
```

```python
for col in train_df[c_below_fifty]:
    replace_missing_mode(train_df, col)
```

## Model Prep: Create our features matrix (`X`) and target vector (`y`)
Let's use the numerical features that have high correlation to saleprice. 

```python
initial = ['overall_qual', 'year_built', 'year_remod/add', 'mas_vnr_area',
       'exter_qual', 'bsmt_qual', 'total_bsmt_sf', '1st_flr_sf', 'gr_liv_area',
       'full_bath', 'kitchen_qual', 'totrms_abvgrd', 'garage_yr_blt',
       'garage_finish', 'garage_cars', 'garage_area']

X = train_df[initial]
y = train_df['saleprice']
```

## Model Prep: Train/test split

```python
# setting the random state to 10 
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=10)
```

## Model Prep: Scaling

```python
# we will scale the features so that they will be treated in the same range
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)

#dont ss.fit(X-test) as the data will be treated differently. Only need to scale once and only on the train data
X_val_sc = ss.transform(X_val)
```

## Model Prep: Instantiate our models
---

We'll be evaluating `LinearRegression`, `LassoCV` and `RidgeCV`.

```python
lr = LinearRegression()

lasso = LassoCV(n_alphas=200)

ridge = RidgeCV(alphas=np.linspace(.005, 20, 100))
```

```python
# Generate the cross_val score for linear regression fitted to train data
lr_scores = cross_val_score(lr, X_train_sc, y_train, cv=5)
# use the scaled data, good to store the scaled data under a new name
lr_scores.mean()
```

```python
# Generate the cross_val score for Lasso CV (i.e. with optimal alpha value) fitted to train data
lasso_scores = cross_val_score(lasso, X_train_sc, y_train, cv=5)
lasso_scores.mean()
```

```python
# Generate the cross_val score for Ridge CV (i.e. with optimal alpha value) fitted to train data
ridge_scores = cross_val_score(ridge, X_train_sc, y_train, cv=5)
ridge_scores.mean()
```

Out of all the models, `RidgeCV` has the highest cross val score of ~ 0.8612, so we will fit our model using that.

```python
ridge.fit(X_train_sc, y_train)
```

```python
ridge.alpha_
# we print this out to check if the alpha is not the start or end value
# finding the optimized alpha value
# start with a wide range scan (0.00005, 1000, 100)
# then print the optimal alpha value 
# then narrow the range down
```

```python
ridge.score(X_train_sc, y_train)
```

```python
ridge.score(X_val_sc, y_val)
```

```python
rmse_train = mean_squared_error(y_train, ridge.predict(X_train_sc), squared = False)
rmse_train
```

```python
rmse_val = mean_squared_error(y_val, ridge.predict(X_val_sc), squared = False)
rmse_val
```

```python
# Create df for storing coefficients and their corresponding variable names
coefficient_df = pd.DataFrame()

# Storing all the column names of predictors
coefficient_df['Column_Name'] = X_train.columns

# Store coefficients of predictors
coefficient_df['Coefficient_Value'] = pd.Series(ridge.coef_)

#sns.barplot(x='Id', y="Speed", data=a, palette=colors, order=result['Id'])
# Preview variables' coefficient values
# The higher the values, the stronger the variables
print(coefficient_df)

# Plot the coefficent values
plt.rcParams['figure.figsize'] = (15,6)
#sns.barplot(x='Column_Name', y="Coefficient_Value", data=train_df, order= u)
plt.bar(coefficient_df["Column_Name"], coefficient_df['Coefficient_Value'])
plt.title('Coefficient of selected features')
plt.xlabel('Features')
plt.ylabel('Coefficient')
plt.xticks(rotation=90)

#sort the visual by absolute value of coefficient
# also orient the labels by 90
# create a column with the absolute coefficient (making negative positive)
# .abs() for absolute 
# sort it in descending

# L2 Regularization has reduced the coefficient values from what has been observed in Linear Regression
# Note: Regularization does not always improve model performance as seen in this case where Linear outperforms Lasso
```

We can identify that some of the features has a weak coefficient influence on the saleprice, such as year built, total rooms above ground, garage year built and garage cars.

After running our initial model, it appears that the model is overfitting. This means that the model is currently collecting too much noise and we can improve this by decreasing the model complexity. We will use this model's R2 score as a baseline.

## Conclusion and Recommendations

Using the Ridge Regression model with a R2 score of 0.83 with a RMSE error of 29196 (scored on Kaggle), we have predicted the core feautres that affect sale price. The top core features that had a high coefficient values on saleprice are overall quality, exterial quality, bsmt quality, ground living area, kitchen quality and garage area. An increase in overall quality, exterial quality, basement quality, ground living area and kitchen quality has a positive influence on increasing the sale price. Real Sky should focus on funneling more of their expenses on improving this features in their next development project to increase the saleprice of their houses. They should also advertise this features. Real Sky should also seek to procure land in areas such as Stone Brooke, Northridge Heights and Northridge neighborhoods for higher saleprice of new houses.

## Limitations
There is insufficient data to fully analyse factors that affect the sale price. Such data like proximity to amenities (i.e distance to malls, schools and public transport), recession and political background can affect the saleprice negatively or positively. 
There was also high amounts of missing data found in the dataset, which were filled with median and mode. This might skew the predictive results of the features.
